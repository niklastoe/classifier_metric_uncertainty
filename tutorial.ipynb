# Tutorial

This tutorial gives a brief overview of the most important parts of the code that allow you to determine and visualize metric uncertainty in your own workflows.

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

from __init__ import ConfusionMatrixAnalyser, bayes_laplace_prior
```

# Input confusion matrix

You need to adapt the entries here to your confusion matrix of interest.

```python
cm = {'TP': 5, 'FN': 2, 'TN': 3, 'FP': 1}
cm = pd.Series(cm)
```

Select the priors

```python
bayeslaplace_priors = {'TNR': [1, 1], 'TPR': [1, 1], 'PREVALENCE': [1, 1]}
```

# Output

```python
cm_analyser = ConfusionMatrixAnalyser(cm, 
                                      priors=bayeslaplace_priors)
```

cm_analyser is an object instance. Its attributes contain all the information that you need.

## posterior samples for confusion probability matrix

```python
cm_analyser.theta_samples
```

## posterior samples for all metrics

```python
cm_analyser.theta_metrics
```

## Metric uncertainties (as defined in the paper, lengths of the 95% credible intervals)

```python
cm_analyser.theta_metric_uncertainty
```

## Interactive visualization of the metric posteriors

You can choose the metric of interest.

```python
cm_analyser.interactive_metric_plot()
```

## Visualize posterior distributions of all metrics in one plot

```python
cm_analyser.plot_all_metrics()
```

## Posterior predictions

These are synthetic confusion matrices generated from the posterior distribution of the confusion probability matrix.

```python
cm_analyser.pp_samples
```

```python
cm_analyser.pp_metrics
```

## Probability to be worse than random guessing

This is equal to $r_{deceptive}$ in the manuscript.

```python
cm_analyser.chance_to_be_harmful()
```
